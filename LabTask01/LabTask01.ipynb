### Part 5: Arithmetic and Data Alignment ###

1. Create a new column that converts fares from USD to Euros.

2. Subtract the average age from every age value.

3. Subtract a custom array from two selected columns in the dataset.

```python
# 1. Convert fare from USD to Euros
df["Fare_Euros"] = df["Fare"] * 0.92

# Check the new column
df[["Fare", "Fare_Euros"]].head(10)
```

```python
# 3. Subtract a custom array from two selected columns
df_age_fare_adjusted = df[["Age", "Fare"]] - [10, 50]

df_age_fare_adjusted.head(10)
```

### Part 7: Mapping and Transformation ###
1. Convert categorical values in the gender column to numeric codes.

2. Create a new column that categorizes age into groups (e.g., child, adult, senior).



```python
# 1. Convert gender/Sex values to numeric codes
df["Sex_Code"] = df["Sex"].map({
    "male": 0,
    "female": 1
})

df[["Name", "Sex", "Sex_Code"]].head(10)
```

```python
# 2. Create age groups
def categorize_age(age):
    if pd.isna(age):
        return "unknown"
    elif age < 18:
        return "child"
    elif age < 60:
        return "adult"
    else:
        return "senior"

df["Age_Group"] = df["Age"].apply(categorize_age)

df[["Name", "Age", "Age_Group"]].head(20)
```

### Part 8: Solving the task without Panda ###
 Choose one task from each of Parts 2, 3, or 5 (3 tasks in total). Try to solve that chosen task without using Pandas, only with pure Python code.

```python
import csv

# Load the CSV file without pandas
with open("titanic.csv", "r", encoding="utf-8") as file:
    reader = csv.DictReader(file)
    data = list(reader)

print(data)
```

```python
# Part 2 task: Select the names column
names = [row["Name"] for row in data]

print("First 10 names:")
print(names[:10])
```

```python
# Part 3 task: Show rows where age is above 60
age_above_60 = []

for row in data:
    if row["Age"] != "":
        age = float(row["Age"])

        if age > 60:
            age_above_60.append(row)

print("\nPassengers above age 60:")
for row in age_above_60:
    print(row["Name"], row["Age"])
```

```python
# Part 5 task: Create a new fare-in-Euros value
for row in data:
    fare_usd = float(row["Fare"])
    fare_euros = fare_usd * 0.92

    row["Fare_Euros"] = fare_euros

print("\nFirst 10 fares in USD and Euros:")
for row in data[:10]:
    print(row["Name"], row["Fare"], row["Fare_Euros"])
```

### Comparison in the execution time ###

```python
import pandas as pd
import csv
import time

# Pandas version
start_time = time.time()

df = pd.read_csv("https://raw.githubusercontent.com/pandas-dev/pandas/main/doc/data/titanic.csv")

names_pandas = df["Name"]

age_above_60_pandas = df[df["Age"] > 60]

df["Fare_Euros"] = df["Fare"] * 0.92

end_time = time.time()

pandas_time = end_time - start_time



# Pure Python version
start_time = time.time()

with open("titanic.csv", "r", encoding="utf-8") as file:
    reader = csv.DictReader(file)
    data = list(reader)

names_python = [row["Name"] for row in data]

age_above_60_python = []

for row in data:
    if row["Age"] != "":
        age = float(row["Age"])

        if age > 60:
            age_above_60_python.append(row)

for row in data:
    fare_usd = float(row["Fare"])
    fare_euros = fare_usd * 0.92

    row["Fare_Euros"] = fare_euros

end_time = time.time()

python_time = end_time - start_time


# Results
print("Pandas execution time:", pandas_time)
print("Pure Python execution time:", python_time)

if pandas_time < python_time:
    print("Pandas was faster.")
else:
    print("Pure Python was faster.")
```

For this small dataset, pure Python was faster.
However, pandas is easier to write, easier to read and read and better for larger data analysis tasks.
